```python
import pandas as pd
import requests
from sqlalchemy import create_engine
```

```python
cost_url = 'https://www.valuepenguin.com/student-loans/average-cost-of-college#state'
```

```python
# Retrieve page with the requests module
response = requests.get(cost_url)
```

```python
cost_tables = pd.read_html(response.text)
cost_tables
```

```python
type(cost_tables)
```

```python
cost_df = cost_tables[1]
cost_df.head()
```

```python
debt_url = 'https://www.valuepenguin.com/average-student-loan-debt#state'
```

```python
# Retrieve page with the requests module
response = requests.get(debt_url)
```

```python
debt_tables = pd.read_html(response.text)
debt_tables
```

```python
type(debt_tables)
```

```python
debt_df = debt_tables[1]
debt_df.head()
```

```python
#Transform into one DataFrame
# Create a filtered dataframe from specific columns
cost_df_cols = ["State", "In-state tuition", "Out-of-state tuition"]
cost_df_transformed= cost_df[cost_df_cols].copy()

# Rename the column headers
cost_df_transformed = cost_df_transformed.rename(columns={"State Name": "State",
                                                           "Tuition- In State": "In-state tuition",
                                                          "Tuition- Out of State":"Out-of-state tuition" })

# Clean the data by dropping duplicates and setting the index
cost_df_transformed.drop_duplicates("State", inplace=True)
cost_df_transformed.set_index("State", inplace=True)

cost_df_transformed.head()
```
